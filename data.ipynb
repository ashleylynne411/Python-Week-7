```python

import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df["species"] = iris.target  # Add species column
print(df.head())  # Show first 5 rows
try:
    df = pd.read_csv("your_file.csv")
    print("CSV file loaded successfully!")
except FileNotFoundError:
    print("Error: File not found. Please check the path.")
except Exception as e:
    print(f"An error occurred: {e}")


print(df.info())  # Shows data types
print(df.isnull().sum())  
df.fillna(df.mean(), inplace=True)  

print(df.describe())  # Gives mean, median, std dev.
grouped_data = df.groupby("species")["sepal length (cm)"].mean() 
print(grouped_data)

import matplotlib.pyplot as plt
import seaborn as sns

df.plot(x="species", y="sepal length (cm)", kind="line") 
plt.title("Flower Data")
plt.xlabel("Species")
plt.ylabel("Sepal length")
plt.show()

df.groupby("species")["petal length (cm)"].mean().plot(kind="bar") 
plt.title("Flower Comparison")
plt.xlabel("Species")
plt.ylabel("Petal length")
plt.show()

sns.histplot(df["petal length (cm)"], bins=20, kde=True) 
plt.title("Histogram showing length of petals")
plt.xlabel("Petal length")
plt.ylabel("Frequency")
plt.show()

sns.scatterplot(x=df["petal length (cm)"], y=df["sepal length (cm)"])  
plt.title("Relationship Between Both Lengths")
plt.xlabel("Petals")
plt.ylabel("Sepals")
plt.show()


```
